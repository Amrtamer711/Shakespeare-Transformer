```python
import torch
import matplotlib.pyplot as plt
```

```python
with open(r'shakespeare_more.txt', 'r', encoding='utf-8') as file:
    text = file.read()
unique_chars = sorted(list(set(text)))
vocab_size = len(unique_chars)
```

```python
itos = {i:s for i, s in enumerate(unique_chars)}
stoi = {s:i for i, s in enumerate(unique_chars)}
encode = lambda x: [stoi[char] for char in x]
decode = lambda x: ''.join([itos[index] for index in x])
data = torch.tensor(encode(text), dtype=torch.long)
n = int(len(data) * 0.9)
data_train = data[:n]
data_test = data[n:]
```

```python
table = torch.ones((vocab_size, vocab_size), dtype=torch.float64)
for i in range(len(data_train)-1):
    table[data_train[i], data_train[i+1]] += 1
```

```python
probs = table / table.sum(axis=1, keepdims=True)
```

```python
train_likelihood = 0
test_likelihood = 0
train_n = len(data_train)-1
test_n = len(data_test)-1
for i in range(train_n):
    train_likelihood += torch.log(probs[data_train[i], data_train[i+1]]).item()
train_loss = -train_likelihood / train_n
print("Training loss is", train_loss)
for i in range(test_n):
    test_likelihood += torch.log(probs[data_test[i], data_test[i+1]]).item()
test_loss = -test_likelihood / test_n
print("Training loss is", test_loss)
```

```python
start = stoi['\ufeff']
last = start
generation = [start]
for i in range(2000):
    next = torch.multinomial(probs[last], num_samples=1, replacement=True).item()
    generation.append(next)
    last = next
sample = decode(generation)
print("Trained sample is:\n", sample)
```
